```python
from llama_index.core import SimpleDirectoryReader
```

```python
documents = SimpleDirectoryReader("/content/data").load_data()
```

```python
documents
```

```python
system_prompt="""
  You are Q&A assitant. Your goal is to answer question as accurately
  as possible based on the instructions and context provided.
"""

## Default format supportable by LLama2
query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
import torch
llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16, "load_in_8bit": True}
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import VectorStoreIndex, ServiceContext
# from llama_index.embeddings.huggingface import HuggingFaceEmbedding

from llama_index.embeddings.huggingface import HuggingFaceEmbedding
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
service_context = ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model

)
```

```python
service_context
```

```python

index = VectorStoreIndex.from_documents(documents, service_context=service_context)
```

```python
index
```

```python
query_engine = index.as_query_engine()
```

```python
query_engine.query("what is fba all about")
```

```python
response = query_engine.query("what is fba all about")
```
