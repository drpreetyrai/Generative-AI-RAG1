```python
from langchain_google_genai import ChatGoogleGenerativeAI

# Create a ChatGoogleGenearativeAI Ojbect and convert systme messages to human readable
llm = ChatGoogleGenerativeAI(
    model="gemini-pro",
    temperature=0.3,
    convert_system_message_to_human=True,

)


# Create a GoogleGenearativeAIEmbedding object for embedding our prompts and doucments
#loading the embedding model from huggingface
embedding_model_name = "sentence-transformers/all-mpnet-base-v2"
# model_kwargs = {"device": "cuda"}
model_kwargs = {"device": "cpu"}
embeddings = HuggingFaceEmbeddings(
  model_name=embedding_model_name,
  model_kwargs=model_kwargs
)
```

<h1>Step2:- Data Loading</h1>

```python
from langchain_community.document_loaders import WebBaseLoader

# Create a WebBaseLoader object with the URL of the blog post to load
loader = WebBaseLoader("https://www.promptingguide.ai/")

# Load the documents from the loader
docs = loader.load()
```

```python
docs
```

```python
print(docs[0])
```

<h1>Step3: Data Splitting/ Creating Chunks</h1>

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Chunk size and overlap
text_splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)

# Store the resulting splits in the `splits` variable.
splits = text_splitter.split_documents(docs)
```

<h1>Step4: Storing Documents</h1>

```python
from langchain_community.vectorstores import Chroma


# passing the hyde embeddings to create and store embeddings
vectorstore = Chroma.from_documents(documents=splits,
                                    collection_name='my-collection',
                                    embedding=embeddings)

# creating Retriever
retriever = vectorstore.as_retriever(search_type="similarity",
                                     search_kwargs={"k": 2})
```

<h1>Step5: Creating a Prompt Template for Generating HyDE </h1>

```python
from langchain.prompts import ChatPromptTemplate
#Creating the Prompt Template
template = """
  for the given question try to generate a hypothetical answer\
  Only generate the answer and nothing else:
  Question: {question}
"""
```

```python
Prompt = ChatPromptTemplate.from_template(template)
query = Prompt.format_prompt(question="What are about prompting")
```
